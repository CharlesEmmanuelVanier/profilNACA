# Tracé simple

* Comme indiqué plus haut, nous n'aurons pas à interagir avec matplotlib lui même, nous aurons juste à interagir avec `pyplot` qui est souvent appelé `plt`.
* Dans l'exemple ci-dessous, on trace des courbes simples de fonctions que l'On obtient à partir de numpy.
* On choisir 500 valeurs de x que l'on injecte dans deux fonctions, un polynôme d'ordre 2 et un polynôme d'ordre 3. Puis on trace tout cela avec `pyplot`.

```python
# Script pour realiser un trace simple

import numpy as np
import matplotlib.pyplot as plt


# quelques parametre pour le graphique
plt.rcParams['font.size'] = 14
plt.rcParams['figure.autolayout'] = True # s'assure que tout rentre dans la figure
plt.rcParams['figure.dpi'] = 100


# donnees a tracer
x = np.linspace(0,10,500)
# Polynome d'ordre 2
y_2 = x**2-2*x+4
# Polynome d'ordre 3
y_3 = 0.1*x**3-1.2*x**2-0.5*x+15


# Bloc de code pour le trace
plt.plot(x,y_2,label='polynome 2nd ordre') # on trace une courbe
plt.plot(x,y_3,label='polynome 3e ordre',linestyle='--') # on trace une autre courbe
plt.xlabel('x') # on définit le label de l'axis x
plt.ylabel('y') # on définit le label de l'axis y
plt.legend() # on spécifie que l'on veut une légende
plt.grid() # on spécifie que l'on veut une grille
plt.title('Mon premier graphique') # on spécifie que l'on veut un titre
plt.savefig('ma_premiere_figure.png') # on spécifie que l'on veut sauver la figure sur le disque en .png
plt.savefig('ma_premiere_figure.pdf') # on spécifie que l'on veut sauver la figure sur le disque en .pdf (pas de perte de résolution)
plt.show() # on affiche la sourbe pour l'objet plt en cours de modification
```

## Application: Vitesse aéronef

```python
import numpy as np
import matplotlib.pyplot as plt

# Données de temps (en secondes)
t = np.array([0, 12, 29, 31, 42, 54])

# Données de vitesse (en m/s)
v = np.array([105, 154, 223, 188, 251, 229])

# 1. Calcul de l'accélération
delta_t = np.diff(t)  # Calcul des variations de temps
delta_v = np.diff(v)  # Calcul des variations de vitess

# Utilisation de la formule de la cinématique pour calculer l'accélération
a = delta_v / delta_t
a = np.insert(a, 0, 0) # On doit ajouter une valeur à a sinon il n'a pas la même longueur que t

# quelques parametre pour le graphique
plt.rcParams['font.size'] = 14
plt.rcParams['figure.autolayout'] = True
plt.rcParams['figure.dpi'] = 100

# Bloc de code pour le trace
plt.plot(t,v,label='vitesse') # on trace une courbe
plt.plot(t,a,label='accélération',linestyle='--') # on trace une autre courbe
plt.xlabel('t') # on définit le label de l'axis x
plt.ylabel('v (m/s) ou a (m2/s)') # on définit le label de l'axis y
plt.legend() # on spécifie que l'on veut une légende
plt.grid() # on spécifie que l'on veut une grille
plt.title('Vitesse et accélération de l\'aéronef') # on spécifie que l'on veut un titre
plt.show() # on affiche la sourbe pour l'objet plt en cours de modification
```
